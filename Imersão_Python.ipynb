**Importação**

```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Imersão-AULA (1).xlsx",sheet_name="Principal")
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel("/content/Imersão-AULA (1).xlsx",sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/Imersão-AULA (1).xlsx",sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/Imersão-AULA (1).xlsx",sheet_name="Chat GPT")
df_chatgpt
```

```python
df_principal=df_principal [['Ativo',	'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal ['Var_pct'] = df_principal ['var_dia_pct'] / 100
df_principal ['valor_inicial'] = df_principal ['valor_final'] / ( 1 + df_principal['Var_pct'])
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal ['Variacao_rs'] = (df_principal['valor_final'] - df_principal ['valor_inicial']) * df_principal ['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica': 'qtd_teorica'}).copy()
df_principal
```

```python
df_principal = df_principal.drop(84)
df_principal = df_principal.drop(85)
df_principal = df_principal.drop(86)
df_principal = df_principal.drop(87)
df_principal = df_principal.drop(88)
df_principal = df_principal.drop(89)
df_principal
```

```python
df_principal = df_principal.drop(81)
df_principal = df_principal.drop(82)
df_principal = df_principal.drop(83)
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Empresa', how='left')
df_principal = df_principal.drop(columns=['Empresa'])
df_principal
```

```python
df_principal['cat-idade'] = df_principal['Idade (anos)'].apply(lambda x: "Mais de 100" if x > 100 else ("Menos de 50" if x < 50 else "Entre 50 e 100"))
df_principal
```

```python
# Calcular o maior valor
maior = df_principal['Variacao_rs'].max()

# Calcular o menor valor
menor = df_principal['Variacao_rs'].min()

# Calcular a média
media = df_principal['Variacao_rs'].mean()

# Calcular a média de lucro
media_lucro = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calcular a média de prejuízo
media_prejuizo = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Imprimir os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Media lucro\tR$ {media_lucro:,.2f}")
print(f"Media prejuízo\tR$ {media_prejuizo:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Área de Atuação')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text = 'Variacao_rs', title='Variação reais por resultado')
fig.show()
```

```python
fig = px.pie(df_analise_segmento, names='Área de Atuação', values='Variacao_rs', title='Variação Reais por Segmento')
fig.show()
```

```python
df_analise_cat_idade = df_principal.groupby('cat-idade')['Variacao_rs'].sum().reset_index()
df_analise_cat_idade
```

```python
fig = px.bar(df_analise_cat_idade, x='cat-idade', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Categoria de Idade')
fig.show()
```
